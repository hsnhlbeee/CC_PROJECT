```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, log1p

# Initialize Spark session
spark = SparkSession.builder.appName("FraudPreprocessing").getOrCreate()

# Load data from GCS
df = spark.read.csv("gs://transactions7/creditcard.csv", header=True, inferSchema=True)
```

```python
df.head()
```

```python
# Feature engineering
df = df.withColumn("log_amount", log1p(col("Amount")))

# Save processed data back to GCS
df.write.csv("gs://transactions7/processed/creditcard_processed.csv", header=True)
```
